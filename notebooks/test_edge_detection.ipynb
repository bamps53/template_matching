```python
import cv2
from matplotlib import pyplot as plt
import numpy as np

template_img = load_image('../data/template_1.jpg')

# 
gray = cv2.cvtColor(template_img, cv2.COLOR_BGR2GRAY)

# 
smoothed = cv2.GaussianBlur(gray, (5, 5), 0)

# 
edges = cv2.Canny(smoothed, threshold1=100, threshold2=200)

# 
plt.imshow(edges)
```
